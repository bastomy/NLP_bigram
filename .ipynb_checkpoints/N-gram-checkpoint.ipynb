#### langkah pertama yang harus dilakukan adalah import library yang dibutuhkan

```python
#import library nltk untuk pemrosesan teks, library re untuk operasi regex dan library Counter untuk menghitung objek
import nltk
import re
from collections import Counter
```

#### langkah kedua adalah mengambil data .txt yang berisi 100 artikel

```python
data_artikel = open("artikel.txt", "r").read().lower() #mengambil teks dari file artikel.txt
```

#### langkah ketiga adalah melakukan Tokenisasi untuk memecah setiap kata

```python
charRemove = re.sub(r'[^\w]', ' ', data_artikel) #menghilangkan tanda baca
tokens = nltk.word_tokenize(charRemove) #tokenisasi menggunakan library nltk
```

```python
print('total kata yang ada pada artikel :', len(tokens))
```

#### langkah keempat adalah mencari frekuensi setiap kata

```python
freq_kata = {}
jmlh_kata = 0
new_token=[]
#membuat perulangan fungsi untuk menghitung kemunculan kata
for token in tokens :
    try:
        int(token) #jika token yang didapat adalah angka
    except ValueError: #jika yang didapat adalah kata
        if token in freq_kata : #cek apakah sudah ada
            freq_kata[token]+=1 #jika sudah makan frekuensi +1
        else :
            freq_kata[token]=1 #jika belum ada maka tambahkan ke dalam list dengan frekuensi = 1
        new_token.append(token)
        jmlh_kata+=1
```

```python
freq_kata
```

#### mencari probabilitas setiap kata menggunakan unigram

```python
#membuat penampung untuk menampung kemungkinan perkata
prob_unigram = {}
for token in new_token :
    prob_unigram[token]=freq_kata[token]/jmlh_kata #probabilitas unigram dari token = frekuensi token / jumlah kata
```

```python
prob_unigram
```

#### pembangunan model bigram

```python
#membangun pasangan bigram dari token yang ada berdasarkan artikel
pasangan_bigram = list(nltk.bigrams(new_token))
```

```python
pasangan_bigram
```

#### mencari frekuensi kata yang telah diolah menggunakan bigram

```python
#mencari frekuensi dari setiap pasangan bigram
frekuensi_bigram = Counter(pasangan_bigram)
```

```python
frekuensi_bigram
```

#### mencari probabilitas pasangan bigram

```python
total_pasangan = len(pasangan_bigram)
prob_bigram = {}
for ps_bigram in pasangan_bigram :
    prob_bigram[ps_bigram] = frekuensi_bigram[ps_bigram]/total_pasangan #probabilitas pasangan bigram = frekuensi pasangan / total pasangan
```

```python
prob_bigram
```

#### fungsi untuk masukan kata yang bertujuan memprediksi kemunculan kata

```python
def suggestionWord(kata):
    prob_terbesar=0;
    prob_big=""
    N = 0
    inv_prob = 1
    for ps_big in prob_bigram : #untuk setiap pasangan bigram
        if (ps_big[0] == kata): #jika pasangan bigram sama dengan kata
            prob = prob_bigram[ps_big] / prob_unigram[kata] #peluang kemunculan pasangan kata = peluang pasangan bigram / peluang unigram dari kata yang dicari
            inv_prob = inv_prob*(1/prob) #untuk perhitungan perplexity
            N += 1
            if(prob > prob_terbesar) : #jika peluang pasangan kata lebih besar dari peluang terbesar
                prob_terbesar = prob #update peluang terbesar
                prob_big = ps_big #update pasangan kata terbaik
    if N==0: #jika kata terakhir tidak ditemukan dalam corpus akan munculkan pesan error
        return "= sugesti kata tidak ditemukan"
    perplex = inv_prob**(1/float(N)) #nilai perpexity dari invers peluang dipangkatkan dengan 1/jumlah tuple
    print(prob_big)
    print("probabilitas terbesar ",prob_terbesar)
    print("Jumlah tuple yang di inputkan: ",N)
    print("Nilai Perplexity: ",perplex)
    return prob_big[1]
```

#### fungsi untuk melakukan evaluasi model dengan perplexity

```python
def kalimat(sentences):
    if sentences!="": #pengecekan kalimat yang dimasukkan apakah empty string atau bukan
        token = nltk.word_tokenize(sentences) #tokenisasi kalimat yang akan dicari sugesti selanjutnya
        last = token[len(token)-1] #mengambil kata terakhir setelah di tokenisasi
        suggestion = suggestionWord(last) #melakukan pencarian sugesti dengan fungsi suggestionWord()
        print(sentences + " "+suggestion) #menampilkan hasil kalimat dengan sugesti yang terbaik berdasarkan bigram
    else:
        print("kalimat tidak boleh kosong") #jika kalimat yang dimasukkan adalah empty string maka akan muncul pesan error
```

```python
kalimat("pada kompetisi")
```

```python
kalimat("kami membawa")
```

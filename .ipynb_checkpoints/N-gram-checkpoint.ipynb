#### langkah pertama yang harus dilakukan adalah import library yang dibutuhkan

```python
import nltk
import re
from collections import Counter
```

#### langkah kedua adalah mengambil data .txt yang berisi 100 artikel

```python
data_artikel = open("artikel.txt", "r").read().lower()
```

```python
data_artikel
```

#### langkah ketiga adalah melakukan Tokenisasi untuk memecah setiap kata

```python
charRemove = re.sub(r'[^\w]', ' ', data_artikel)
tokens = nltk.word_tokenize(charRemove)
```

```python
print('total kata yang ada pada artikel :', len(tokens))
```

#### langkah keempat adalah mencari frekuensi setiap kata

```python
freq_kata = {}
jmlh_kata = 0
new_token=[]
#membuat perulangan fungsi untuk menghitung kemunculan kata
for token in tokens :
    try:
        int(token)
    except ValueError:
        if token in freq_kata :
            freq_kata[token]+=1
        else :
            freq_kata[token]=1
        new_token.append(token)
        jmlh_kata+=1
```

```python
freq_kata
```

#### mencari probabilitas setiap kata menggunakan unigram

```python
#membuat penampung untuk menampung kemungkinan perkata
prob_unigram = {}
for token in new_token :
    prob_unigram[token]=freq_kata[token]/jmlh_kata
```

```python
prob_unigram
```

#### pembangunan model bigram

```python
pasangan_bigram = list(nltk.bigrams(new_token))
```

```python
pasangan_bigram
```

#### mencari frekuensi kata yang telah diolah menggunakan bigram

```python
frekuensi_bigram = Counter(pasangan_bigram)
```

```python
frekuensi_bigram
```

#### mencari probabilitas pasangan bigram

```python
total_pasangan = len(pasangan_bigram)
prob_bigram = {}
for ps_bigram in pasangan_bigram :
    prob_bigram[ps_bigram] = frekuensi_bigram[ps_bigram]/total_pasangan
```

```python
prob_bigram
```
